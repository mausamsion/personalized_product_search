```python
import pandas as pd
import gzip
import json
import numpy as np
from tqdm import tqdm
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def parse_text(path):
    text = []
    with gzip.open(path, 'rt') as f:
        text = [line[:-1] for line in f]
    return text
```

```python
query_text = parse_text('apsd/reviews_Cell_Phones_and_Accessories_5.json.gz/query_text.txt.gz')
```
